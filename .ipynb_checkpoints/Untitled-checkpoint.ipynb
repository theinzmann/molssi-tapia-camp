## Plotting Lesson

```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_path = os.path.join('distance_data_headers.csv')
errythang = np.genfromtxt(file_path, delimiter=',', dtype='unicode')
```

```python
data = errythang[1:,:]
header = errythang[0,:]
data = data.astype(np.float)
```

```python
for i in range(1,5):
    plt.plot(data[:,0], data[:,i], linewidth=1, label=header[i])

plt.xlabel('Frame'); plt.ylabel('Distance (Angstrom)')
plt.title('MD Simulation')
plt.legend(); plt.savefig('md_simulaton.png', dpi=300)
plt.show()
```

```python
for i in range(1,5):
    plt.subplot(2,2,i)
    plt.plot(data[:,0], data[:,i], linewidth=1, label=header[i])
    plt.xlabel('Frame'); plt.ylabel('Distance (Angstrom)')
    plt.title('MD Simulation')
    plt.legend(loc='best');

plt.tight_layout()
plt.show()
```

```python
for i in range(1,5):
    plt.figure(i)
    plt.plot(data[:,0], data[:,i], linewidth=1, label=header[i])
    plt.xlabel('Frame'); plt.ylabel('Distance (Angstrom)')
    plt.title('MD Simulation')
    plt.legend(loc='upper right'); plt.savefig('md_simulaton{}.png'.format(i), dpi=300)
    plt.show()
```
